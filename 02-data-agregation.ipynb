Build the T1 dataset

```python
# Step 1: Get results with start position (grid) and finish position
results = dataframes['results.csv'].copy()

# Filter out DNF/DNS (only keep drivers who finished with a valid position)
results = results[results['positionOrder'].notna() & (results['positionOrder'] > 0)]
results = results[results['grid'] > 0]  # Valid starting position

# Calculate overtakes: positive = gained positions, negative = lost positions
# Overtakes = grid position - finish position (higher grid number = worse starting position)
results['overtakes'] = results['grid'] - results['positionOrder']

# Create a new clean table with only the required columns
overtakes_table = results[['raceId', 'driverId', 'grid', 'positionOrder', 'overtakes']].copy()

# Rename columns for clarity
overtakes_table.columns = ['raceId', 'driverId', 'start_pos', 'end_pos', 'overtakes']

# Reorder columns as requested
overtakes_table = overtakes_table[['raceId', 'driverId', 'overtakes', 'start_pos', 'end_pos']]

overtakes_table.head(20)
```

```python
# Step 2: Calculate average lap time per driver per race
lap_times = dataframes['lap_times.csv'].copy()
grouped = lap_times.groupby(['raceId', 'driverId'])
avg_lap_times = grouped['milliseconds'].mean().reset_index()
avg_lap_times.columns = ['raceId', 'driverId', 'avg_lap_time_ms']

# Merge with overtakes_table
overtakes_table = overtakes_table.merge(avg_lap_times, on=['raceId', 'driverId'], how='left')

print(f"Overtakes table shape: {overtakes_table.shape}")
print(f"\nColumns: {list(overtakes_table.columns)}")
overtakes_table.head(10)
```

```python
# Step 3: Add circuit (track) information
# First, get circuitId from races
races_circuit = dataframes['races.csv'][['raceId', 'circuitId']].copy()
overtakes_table = overtakes_table.merge(races_circuit, on='raceId', how='left')

# Then merge with circuit names
circuits = dataframes['circuits.csv'][['circuitId', 'name']].copy()
circuits.columns = ['circuitId', 'circuit_name']
overtakes_table = overtakes_table.merge(circuits, on='circuitId', how='left')

print(f"Overtakes table shape: {overtakes_table.shape}")
print(f"\nColumns: {list(overtakes_table.columns)}")
overtakes_table.head(10)
```

```python
# Step 4: Add constructor (team) information
# First, get constructorId from results
results_constructor = dataframes['results.csv'][['raceId', 'driverId', 'constructorId']].copy()
overtakes_table = overtakes_table.merge(results_constructor, on=['raceId', 'driverId'], how='left')

# Then merge with constructor names
constructors = dataframes['constructors.csv'][['constructorId', 'name']].copy()
constructors.columns = ['constructorId', 'team_name']
overtakes_table = overtakes_table.merge(constructors, on='constructorId', how='left')

print(f"Overtakes table shape: {overtakes_table.shape}")
print(f"\nColumns: {list(overtakes_table.columns)}")
overtakes_table.head(10)
```

```python
# Step 5: Add driver names
drivers = dataframes['drivers.csv'][['driverId', 'forename', 'surname']].copy()
drivers['driver_name'] = drivers['forename'] + ' ' + drivers['surname']
drivers = drivers[['driverId', 'driver_name']]
overtakes_table = overtakes_table.merge(drivers, on='driverId', how='left')

print(f"Overtakes table shape: {overtakes_table.shape}")
print(f"\nColumns: {list(overtakes_table.columns)}")
overtakes_table.head(10)
```

```python
# Save the overtakes_table to CSV
output_dir = 'dataset'
output_file = 'dataset/t1/t1.csv'

# Create directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Save to CSV
overtakes_table.to_csv(output_file, index=False)

print(f"Table saved to: {output_file}")
print(f"Shape: {overtakes_table.shape}")
print(f"Columns: {list(overtakes_table.columns)}")
```
