```python
import kagglehub
import shutil
import os

cached_path = kagglehub.dataset_download("rohanrao/formula-1-world-championship-1950-2020")

destination_folder = "./dataset/kaggle"

if os.path.exists(destination_folder):
    shutil.rmtree(destination_folder)
    
shutil.copytree(cached_path, destination_folder)

print(f"Success! Files moved to: {destination_folder}")
```
